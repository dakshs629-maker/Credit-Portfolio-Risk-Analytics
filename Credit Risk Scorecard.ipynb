```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
cat_cols= data.select_dtypes(include= 'object').columns

for col in cat_cols:
    data[col]= le.fit_transform(data[col].astype(str))

data.dtypes.value_counts()
```

```python
def calc_iv(df, feature, target):
    lst= []
    for val in df[feature].unique():
        lst.append({
            'Value' : val,
            'All' : df[df[feature]== val].shape[0],
            'Bad' : df[(df[feature]== val) & (df[target]== 1)].shape[0]
        })

    dset = pd.DataFrame(lst)
    dset['Good']= dset['All'] - dset['Bad']
    dset['Distr_Good']= dset['Good'] / dset['Good'].sum()
    dset['Distr_Bad']= dset['Bad'] / dset['Bad'].sum()
    dset['Distr_Good']= dset['Distr_Good'].replace(0, 0.0001)
    dset['Distr_Bad']= dset['Distr_Bad'].replace(0, 0.0001)
    dset['WoE']= np.log(dset['Distr_Good'] / dset['Distr_Bad'])
    dset['IV']= (dset['Distr_Good'] - dset['Distr_Bad']) * dset['WoE']
    return dset['IV'].sum()

iv_scores= {col: calc_iv(data, col, 'TARGET') for col in data.columns if col != 'TARGET'}
iv_df= pd.DataFrame.from_dict(iv_scores, orient= 'index', columns= ['IV']).sort_values('IV', ascending= False)
print(iv_df.head(20))
```

```python
iv_df.to_csv('iv_scores.csv')

selected_features= iv_df[iv_df['IV'] > 0.02].index.tolist()
selected_features.remove('SK_ID_CURR')
print(len(selected_features))
print(selected_features)
```

```python
from sklearn.model_selection import train_test_split

X= data[selected_features]
y= data['TARGET']

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25 , stratify=y, random_state= 42)

print(X_train.shape, X_test.shape)
print(y_train.value_counts(normalize= True))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

scaler= StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)

lr= LogisticRegression(class_weight= 'balanced', max_iter= 2000, random_state= 42)
lr.fit(X_train_scaled, y_train)

y_pred_proba= lr.predict_proba(X_test_scaled)[:, 1]
print('AUC-ROC:', roc_auc_score(y_test, y_pred_proba))
```

```python
from xgboost import XGBClassifier

xgb= XGBClassifier(
    n_estimators= 300,
    max_depth= 5,
    learning_rate= 0.05,
    scale_pos_weight= 11,
    random_state= 42,
    eval_metric= 'auc'
)

xgb.fit(X_train, y_train)
y_pred_xgb= xgb.predict_proba(X_test)[:,1]
print('XGB AUC-ROC:', roc_auc_score(y_test, y_pred_xgb))
```

```python
import shap

explainer= shap.TreeExplainer(xgb)
shap_values= explainer.shap_values(X_test[:1000])

shap.summary_plot(shap_values, X_test[:1000], feature_names= selected_features)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve

fpr, tpr, thresholds= roc_curve(y_test, y_pred_xgb)
ks_stat= max(tpr - fpr)
gini= 2 *roc_auc_score(y_test, y_pred_xgb) - 1

y_pred_class= (y_pred_xgb >= 0.5).astype(int)

print(f'KS Statistic: {ks_stat: .4f}')
print(f'Gini Coefficient: {gini: .4f}')
print(f'AUC-ROC: {roc_auc_score(y_test, y_pred_xgb): .4f}')

print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred_class))

print('\nClassification Report:')
print(classification_report(y_test, y_pred_class))
```
